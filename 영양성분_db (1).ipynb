```python
food_db_name = "통합 식품영양성분DB_음식_20240211.xlsx"
food_db=pd.read_excel(food_db_name)
```

```python
# 일반적인 종이컵은 6.5oz(192ml)이다. 
# 1온스 = 28.35g, 종이컵 용량은 184.3g으로 환산 가능
```

```python
food_db.describe()
```

```python
food_db['종이컵_단위'] = food_db.apply(lambda row: row['1회제공량'] / 184.3 if row['내용량_단위'] == 'g' else row['1회제공량'] / 192, axis=1)
```

```python
food_df = food_db[['식품명', '1회제공량','내용량_단위','에너지(㎉)','단백질(g)', '지방(g)','탄수화물(g)', '총당류(g)','나트륨(㎎)', '콜레스테롤(㎎)', '총 식이섬유(g)', '종이컵_단위']]
food_df
```

```python
def standardize_nutrient_amount(food_df):
    food_df['1회제공량'] = pd.to_numeric(food_df['1회제공량'], errors='coerce') 
    nutrient_columns = ['에너지(㎉)', '단백질(g)', '지방(g)', '탄수화물(g)', '총당류(g)', '나트륨(㎎)', '콜레스테롤(㎎)', '총 식이섬유(g)']
    for col in nutrient_columns:
        food_df[col] = pd.to_numeric(food_df[col], errors='coerce')  / (food_df['1회제공량'] / 100)
    return food_df
```

```python
food_df = standardize_nutrient_amount(food_df)
food_df
```

```python
nutrient_content = np.full(len(food_df), 100)
food_df['영양성분함량기준량'] = nutrient_content
food_df.head()
```

```python
def parse_weight_volume(value):
    if isinstance(value, str):
        pattern = r'(\d+)([a-zA-Z]+)'
        match = re.match(pattern, value)
        if match:
            amount = match.group(1)
            unit = match.group(2)
            return amount, unit
        else:
            return None, None
    else:
        return None, None
        
food_db2[['내용량_제공량', '내용량_단위']] = food_db2['1회제공량'].apply(lambda x: pd.Series(parse_weight_volume(x)))

print(food_db2)
```
