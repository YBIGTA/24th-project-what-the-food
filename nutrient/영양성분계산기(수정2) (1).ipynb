```python
class HealthCalculator:
    def __init__(self):
        pass

    def calculate_daily_calories(self, activity_level):
        if activity_level == "비활동적":
            PA = 1.0
        elif activity_level == "저활동적":
            PA = 1.12
        elif activity_level == "활동적":
            PA = 1.27
        else:
            PA = 1.45
        return PA

    def calculate_bmi(self, weight, height):
        bmi = weight / (height ** 2)
        bmi_range = self.calculate_bmi_range(bmi)
        return bmi, bmi_range

    def calculate_bmi_range(self, bmi):
        if bmi < 18.5:
            res = "저체중"
        elif bmi < 23:
            res = "정상"
        elif bmi < 25:
            res = "과체중"
        elif bmi < 30:
            res = "1단계 비만"
        elif bmi < 35:
            res = "2단계 비만"
        else:
            res = "3단계 비만"
        return res
        
    def calculate_EER(self, weight, height, sex, age, PA):
        if sex == "F":
            EER = 354-(6.91*age)+PA*(9.36*weight+726*height)
        else:
            EER = 622-(9.53*age)+PA*(15.91*weight+539.6*height)
        return EER

    def calculate_EER_diet(self, weight, goal_weight, goal_week, EER):
        if weight == goal_weight:
            EER_D = EER
        else:
            EER_D = EER - (weight - goal_weight) * 7000 / goal_week
        return EER_D

    def calculate_health_info(self):
        while True:
            try:
                weight = float(input("몸무게(kg)를 입력해주세요: "))
                if not (weight > 0):
                    raise ValueError
                break
            except ValueError:
                print("몸무게를 숫자로 다시 입력해주세요.")
    
        while True:
            try:
                height = float(input("키(cm)를 입력해주세요: ")) / 100
                if not (height > 0):
                    raise ValueError
                break
            except ValueError:
                print("키를 숫자로 다시 입력해주세요.")
    
        while True:
            try:
                age = int(input("나이(세)를 입력해주세요: "))
                if not (age > 0):
                    raise ValueError
                break
            except ValueError:
                print("나이를 숫자로 다시 입력해주세요.")
    
        while True:
            try:
                sex = input("성별을 입력해주세요(F/M): ")
                if not (sex.upper() == 'F' or sex.upper() == 'M'):
                    raise ValueError
                break
            except ValueError:
                print("성별을 'F' 또는 'M'으로 다시 입력해주세요.")
    
        while True:
            activity_level = input("활동성을 입력해주세요(비활동적, 저활동적, 활동적, 매우 활동적): ")
            if activity_level.lower() in ['비활동적', '저활동적', '활동적', '매우 활동적']:
                break
            else:
                print("활동성을 다시 입력해주세요.")
    
        while True:
            try:
                goal_weight = float(input("목표 몸무게(kg)를 입력해주세요: "))
                if not (goal_weight > 0):
                    raise ValueError
                break
            except ValueError:
                print("목표 몸무게를 숫자로 다시 입력해주세요.")
    
        while True:
            try:
                goal_week = int(input("목표 몸무게 달성 기간(일)을 입력해주세요: "))
                if not (goal_week > 0):
                    raise ValueError
                break
            except ValueError:
                print("목표 몸무게 달성 기간을 숫자로 다시 입력해주세요.")

        PA = self.calculate_daily_calories(activity_level)
        bmi, bmi_range = self.calculate_bmi(weight, height)
        EER = self.calculate_EER(weight, height, sex, age, PA)
        EER_D = self.calculate_EER_diet(weight, goal_weight, goal_week, EER)
    
        print("BMI:", bmi)
        print("비만도:", bmi_range)
        print("일일 권장 섭취 칼로리량:", EER_D)
    
        return {"몸무게":weight, "키": height, "나이": age, "성별":sex, "활동성": activity_level, "목표 몸무게": goal_weight, 
                "다이어트 기간": goal_week, "PA": PA, "BMI":bmi, "비만정도":bmi_range, "EER": EER, "EER_D": EER_D}

# 클래스의 인스턴스 생성
calculator = HealthCalculator()
user_info = calculator.calculate_health_info()
print(user_info)
```

```python
class NutritionCalculator:
    def __init__(self, EER, sex, num_meals, food_df):
        self.daily_nutrient = self.daily_nutrient_intake(EER, sex)
        self.total_nutrition_consumption = {
            '탄수화물(g)': 0,
            '단백질(g)': 0,
            '지방(g)': 0,
            '나트륨(mg)': 0,
            '콜레스테롤(mg)': 0,
            '당류(g)': 0
        }
        self.num_meals = num_meals
        self.meal_limits = self.calculate_meal_limits()
        self.food_df = food_df

    def daily_nutrient_intake(self, EER, sex):
        carb_calories = EER * 0.5
        protein_calories = EER * 0.3
        fat_calories = EER * 0.2

        carb_grams = carb_calories / 4
        protein_grams = protein_calories / 4
        fat_grams = fat_calories / 9
        
        cholesterol_limit = 300
        sugar_limit = 37.5 if sex == 'M' else 25
        sodium_limit = 2400

        return {
            '탄수화물(g)': carb_grams,
            '단백질(g)': protein_grams,
            '지방(g)': fat_grams,
            '나트륨(mg)': sodium_limit,
            '콜레스테롤(mg)': cholesterol_limit,
            '당류(g)': sugar_limit
        }

    def calculate_meal_limits(self):
        meal_limits = {}
        for nutrient, limit in self.daily_nutrient.items():
            meal_limits[nutrient] = limit / self.num_meals
        return meal_limits

    def calculate_nutrition_consumption(self, food_name):
        food_row = self.food_df.loc[self.food_df['식품명'] == food_name]
        
        if food_row.empty:
            print(f"Error: '{food_name}'에 해당하는 음식이 데이터프레임에 없습니다.")
            return None
        
        carb_consume = food_row['탄수화물(g)'].sum()
        protein_consume = food_row['단백질(g)'].sum()
        fat_consume = food_row['지방(g)'].sum()
        sodium_consume = food_row['나트륨(mg)'].sum()
        cholesterol_consume = food_row['콜레스테롤(mg)'].sum()
        sugar_consume = food_row['당류(g)'].sum()

        self.total_nutrition_consumption['탄수화물(g)'] += carb_consume
        self.total_nutrition_consumption['단백질(g)'] += protein_consume
        self.total_nutrition_consumption['지방(g)'] += fat_consume
        self.total_nutrition_consumption['나트륨(mg)'] += sodium_consume
        self.total_nutrition_consumption['콜레스테롤(mg)'] += cholesterol_consume
        self.total_nutrition_consumption['당류(g)'] += sugar_consume
        
        consumed_ratio = {}
        for nutrient, limit in self.daily_nutrient.items():
            consumed_ratio[nutrient] = f"{self.total_nutrition_consumption[nutrient]}/{limit}"
        
        for nutrient, ratio in consumed_ratio.items():
            if self.total_nutrition_consumption[nutrient] > self.daily_nutrient[nutrient]:
                print(f"Warning: {food_name}은(는) {nutrient}을(를) {ratio} 섭취하고 있습니다.")
        
        return self.total_nutrition_consumption, consumed_ratio

calculator = NutritionCalculator(user_info["EER"], user_info["성별"], float(input("오늘 하루 식사횟수를 입력해주세요: ")), food)

while True:
    food_name = input("음식 이름을 입력하세요 (종료하려면 '끝'을 입력하세요): ")
    if food_name == "끝":
        break
    total_consumption, consumption_ratio = calculator.calculate_nutrition_consumption(food_name)
    print("총 영양소 소비량:", total_consumption)
    print("영양소 소비량 비율:", consumption_ratio)
    print("\n")
```

```python
class NutritionCalculator_2:
    def __init__(self, EER, sex, num_meals, food_df):
        self.daily_nutrient = self.daily_nutrient_intake(EER, sex)
        self.total_nutrition_consumption = {
            '탄수화물(g)': 0,
            '단백질(g)': 0,
            '지방(g)': 0,
            '나트륨(mg)': 0,
            '콜레스테롤(mg)': 0,
            '당류(g)': 0
        }
        self.num_meals = num_meals
        self.meal_limits = self.calculate_meal_limits()
        self.food_df = food_df

    def daily_nutrient_intake(self, EER, sex):
        carb_calories = EER * 0.5
        protein_calories = EER * 0.3
        fat_calories = EER * 0.2

        carb_grams = carb_calories / 4
        protein_grams = protein_calories / 4
        fat_grams = fat_calories / 9
        
        cholesterol_limit = 300
        sugar_limit = 37.5 if sex == 'M' else 25
        sodium_limit = 2400

        return {
            '탄수화물(g)': carb_grams,
            '단백질(g)': protein_grams,
            '지방(g)': fat_grams,
            '나트륨(mg)': sodium_limit,
            '콜레스테롤(mg)': cholesterol_limit,
            '당류(g)': sugar_limit
        }

    def calculate_meal_limits(self):
        meal_limits = {}
        for nutrient, limit in self.daily_nutrient.items():
            meal_limits[nutrient] = limit / self.num_meals
        return meal_limits

    def calculate_nutrition_consumption(self, food_name):
        food_row = self.food_df.loc[self.food_df['식품명'] == food_name]
        
        if food_row.empty:
            print(f"Error: '{food_name}'에 해당하는 음식이 데이터프레임에 없습니다.")
            return None
        
        carb_consume = food_row['탄수화물(g)'].sum()
        protein_consume = food_row['단백질(g)'].sum()
        fat_consume = food_row['지방(g)'].sum()
        sodium_consume = food_row['나트륨(mg)'].sum()
        cholesterol_consume = food_row['콜레스테롤(mg)'].sum()
        sugar_consume = food_row['당류(g)'].sum()

        self.total_nutrition_consumption['탄수화물(g)'] += carb_consume
        self.total_nutrition_consumption['단백질(g)'] += protein_consume
        self.total_nutrition_consumption['지방(g)'] += fat_consume
        self.total_nutrition_consumption['나트륨(mg)'] += sodium_consume
        self.total_nutrition_consumption['콜레스테롤(mg)'] += cholesterol_consume
        self.total_nutrition_consumption['당류(g)'] += sugar_consume
        
        consumed_ratio = {}
        for nutrient, limit in self.daily_nutrient.items():
            consumed_ratio[nutrient] = f"{self.total_nutrition_consumption[nutrient]}/{limit}"
        
        for nutrient, ratio in consumed_ratio.items():
            if self.total_nutrition_consumption[nutrient] > self.daily_nutrient[nutrient]:
                print(f"Warning: {food_name}은(는) {nutrient}을(를) {ratio} 섭취하고 있습니다.")
        
        return self.total_nutrition_consumption, consumed_ratio
        
    def propose_food_consumption(self, food_names):
        consumption_proposals = {}
        valid_food_names = [name.strip() for name in food_names.split(",")]
    
        for food_name in valid_food_names:
            food_row = self.food_df.loc[self.food_df['식품명'] == food_name]
            if food_row.empty:
                print(f"Error: '{food_name}'에 해당하는 음식이 데이터프레임에 없습니다.")
                continue
            
            for nutrient in self.daily_nutrient.keys():
                if nutrient in ['식품명', '1회제공량', '내용량_단위', '영양성분함량기준량']:
                    continue
                amount = food_row[nutrient].iloc[0]
                try:
                    amount = float(amount)
                    consumption_proposals[nutrient] = (amount / 100) * self.meal_limits[nutrient]
                except ValueError:
                    print(f"Error: '{nutrient}'에 대한 데이터가 숫자가 아닙니다.")
    
        return consumption_proposals
    
calculator = NutritionCalculator_2(user_info["EER"], user_info["성별"], float(input("오늘 하루 식사횟수를 입력해주세요: ")), food)

while True:
    food_names = input("음식 이름을 입력하세요 (끝을 입력하면 종료): ")
    if food_names == "끝":
        break
    consumption_proposals = calculator.propose_food_consumption(food_names)
    print(f"{food_names}에 대한 섭취량 제안:")
    for nutrient, amount in consumption_proposals.items():
        print(f"{nutrient}: {amount:.2f}g 제안")
    print()
```
